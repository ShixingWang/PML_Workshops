 ## What is Image Segmentation?
 Here is a definition quoted from Wikipedia:
 > In computer vision, image segmentation is the process of partitioning a digital image into multiple segments (sets of pixels, also known as super-pixels).

 An image is a container full of data, in the form of pixels. The neat shape of an image makes it easy to process with tensor manipulation mathods, but this comes at a cost that the pixels not belonging to the objects of our interest also appear in the image. Image segmentation distinguishes the objects of our interest and "segments" objects and backgrounds into different super-pixels.

 Image segmentation attracted people's attention much earlier than the machine leanring got popular, so there have been effective methods without utilizing fancy machine learning methods, but AI really brings this topic to a higher level. For the same methods, there are different implements on different libraries. So this notebook will be organized into a two-level hierachy. The upper level introduces these segmentation methods, while the lower level talks about the codes calling the functions that really work. Here is the list:

 * Supervised (manual) segmentation
   - scikit-image
 * Unsupervised segmentation
   - scikit-image
 * Semantic segmentation
   - TensorFlow
